6.finding and replacing outliers

```python
sns.boxplot(df.CreditScore)
```

```python
q3=df.CreditScore.quantile(0.75)
q1=df.CreditScore.quantile(0.25)
iqr=q3-q1
```

```python
up_limit=q3+1.5*iqr
lw_limit=q1-1.5*iqr
```

```python
df.median()
```

```python
df['CreditScore']=np.where(df['CreditScore']<lw_limit,652,df['CreditScore'])
```
